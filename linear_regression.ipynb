```python
import pandas as pd 
from sklearn import linear_model
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv('canada_per_capita_income.csv')
data
```

```python
plt.bar(data.year,data.income,color='blue')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (In US$)')
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[['year']],data.income)
```

```python
reg.predict([[2020]])
```

```python
years=pd.Series([2017,2018,2019,2020,2021,2022,2023,2024])
```

```python
df=pd.DataFrame(years,columns=['year'])
```

```python
income=reg.predict(df)
df['income']=income
```

```python
df
```

```python
df.to_csv('predicted_income.csv',index=False)
```

```python
new_df=pd.concat([data,df],axis=0)
```

```python
new_df
```

```python
new_df.to_csv('predicted_income.csv',index=False)
```
